```python
# GroupDNA - WhatsApp Chat Analyzer
```

```python
# Import Libraries

import numpy as np
from datetime import datetime
```

```python
# Upload Dataset

filename = "hostel_bois.txt"
```

```python
# Read the File

with open(filename, "r", encoding="utf-8") as file:
    lines = file.readlines()

print("Total Lines:", len(lines))
```

```python
# Create Variables

messages = []

system_count = 0
media_count = 0
deleted_count = 0
```

```python
# Parse the Chat

for line in lines:

    line = line.strip()

    # Skip empty lines
    if line == "":
        continue

    if (
        len(line) < 8 or
        line[2] != "/" or
        line[5] != "/"
    ):

        if len(messages) > 0:
            messages[-1]["text"] += " " + line

        continue

    # Split timestamp and remaining text
    try:

        timestamp, remaining = line.split(" - ", 1)

    except:

        continue

    # System message check
    if ": " not in remaining:

        system_count += 1

        continue

    sender, text = remaining.split(": ", 1)

    # Count media
    if text == "<Media omitted>":

        media_count += 1

    # Count deleted
    if text == "This message was deleted":

        deleted_count += 1

    # Store message
    messages.append({

        "timestamp": timestamp,

        "sender": sender,

        "text": text

    })
```

```python
# Feature 5 - Word Analysis
```

```python
# Stop Words

stop_words = {
    "the","is","a","an","to","and","of","in","for",
    "on","at","it","this","that","i","you","me",
    "my","we","our","your","are","was","were","be",
    "am","do","did","have","has","had","will","shall",
    "can","could","would","should","with","from","as",
    "or","if","but","so","not","yes","no","ok","okay"
}
```

```python
# Remove Punctuation & Count Words

word_count = {}

punctuation = ".,!?;:'\"()[]{}<>/-_@#$%^&*=+|`~"

for msg in messages:

    text = msg["text"].lower()

    # Skip media & deleted messages
    if text == "<media omitted>":
        continue

    if text == "this message was deleted":
        continue

    words = text.split()

    for word in words:

        word = word.strip(punctuation)

        if word == "":
            continue

        if word in stop_words:
            continue

        if word in word_count:

            word_count[word] += 1

        else:

            word_count[word] = 1
```

```python
# Total Unique Words

print("Unique Words :", len(word_count))
```

```python
# Sort Words

sorted_words = sorted(
    word_count.items(),
    key=lambda item: item[1],
    reverse=True
)

print("Top 10 Words\n")

for word, count in sorted_words[:10]:

    print(word, ":", count)
```

```python
# Feature 6 - Response Analysis
```

```python
# Convert Timestamp to Datetime

for msg in messages:

    msg["datetime"] = datetime.strptime(
        msg["timestamp"],
        "%d/%m/%y, %H:%M"
    )

print(messages[0]["datetime"])
```

```python
# Store Response Times

response_times = {}

for person in participants:

    response_times[person] = []
```

```python
# Calculate Response Times

for i in range(1, len(messages)):

    previous = messages[i - 1]
    current = messages[i]

    if previous["sender"] != current["sender"]:

        gap = (
            current["datetime"] -
            previous["datetime"]
        ).total_seconds() / 60

        if gap >= 0:

            response_times[current["sender"]].append(gap)
```

```python
# Average Response Time

average_response = {}

for person in participants:

    times = response_times[person]

    if len(times) > 0:

        average = sum(times) / len(times)

    else:

        average = 0

    average_response[person] = average

print(average_response)
```

```python
# Fastest Replier

fastest_person = ""
fastest_time = float("inf")

for person, avg in average_response.items():

    if avg > 0 and avg < fastest_time:

        fastest_time = avg
        fastest_person = person

print("Fastest Replier")

print(fastest_person)

print(round(fastest_time,2), "minutes")
```

```python
# Slowest Replier

slowest_person = ""
slowest_time = 0

for person, avg in average_response.items():

    if avg > slowest_time:

        slowest_time = avg
        slowest_person = person

print("Slowest Replier")

print(slowest_person)

print(round(slowest_time,2), "minutes")
```

```python
# Print All Response Times

print("="*60)
print("AVERAGE RESPONSE TIMES")
print("="*60)

for person in sorted(participants):

    avg = average_response[person]

    if avg < 60:

        print(f"{person:<10} {avg:.1f} minutes")

    else:

        hours = avg / 60

        print(f"{person:<10} {hours:.2f} hours")
```

```python
# Silent Streak Variables

silent_days = {}

for person in participants:

    silent_days[person] = 0
```

```python
# Longest Silent Streak

for person in participants:

    person_dates = []

    for msg in messages:

        if msg["sender"] == person:

            day = msg["datetime"].date()

            if day not in person_dates:

                person_dates.append(day)

    longest = 0

    for i in range(1, len(person_dates)):

        gap = (person_dates[i] - person_dates[i-1]).days - 1

        if gap > longest:

            longest = gap

    silent_days[person] = longest
```

```python
# Print Silent Streaks

print("="*60)
print("LONGEST SILENT STREAK")
print("="*60)

for person in sorted(participants):

    print(f"{person:<10} {silent_days[person]} days")
```

```python
# Summary

print("="*70)
print("RESPONSE ANALYSIS")
print("="*70)

print()

print("Fastest Replier")

print(f"{fastest_person} ({fastest_time:.1f} minutes)")

print()

print("Slowest Replier")

if slowest_time < 60:

    print(f"{slowest_person} ({slowest_time:.1f} minutes)")

else:

    print(f"{slowest_person} ({slowest_time/60:.2f} hours)")

print()

print("Longest Silent Streaks")

for person in sorted(participants):

    print(f"{person:<10} {silent_days[person]} days")
```

```python
# Feature 7 - Personality Archetypes
```

```python
# Create Score Dictionary

archetypes = {}

for person in participants:

    archetypes[person] = {}
```

```python
# The Spammer

spam_score = {}

for person in participants:

    streaks = []
    streak = 0

    for msg in messages:

        if msg["sender"] == person:

            streak += 1

        else:

            if streak > 0:
                streaks.append(streak)

            streak = 0

    if streak > 0:
        streaks.append(streak)

    if len(streaks) > 0:

        spam_score[person] = sum(streaks) / len(streaks)

    else:

        spam_score[person] = 0

for person in participants:

    archetypes[person]["SPAMMER"] = spam_score[person]
```

```python
# The Night Owl

night_score = {}

for person in participants:

    total = 0
    night = 0

    for msg in messages:

        if msg["sender"] == person:

            total += 1

            hour = msg["datetime"].hour

            if hour >= 23 or hour <= 4:

                night += 1

    if total > 0:

        night_score[person] = (night / total) * 100

    else:

        night_score[person] = 0

for person in participants:

    archetypes[person]["NIGHT OWL"] = night_score[person]
```

```python
# The Story Teller

story_score = {}

for person in participants:

    total_words = 0
    total_messages = 0

    for msg in messages:

        if msg["sender"] == person:

            total_words += len(msg["text"].split())

            total_messages += 1

    if total_messages > 0:

        story_score[person] = total_words / total_messages

    else:

        story_score[person] = 0

for person in participants:

    archetypes[person]["STORYTELLER"] = story_score[person]
```

```python
# The Drama Queen

drama_score = {}

for person in participants:

    total = 0
    caps = 0

    for msg in messages:

        if msg["sender"] == person:

            total += 1

            text = msg["text"]

            if text.isupper() and len(text) > 3:

                caps += 1

    if total > 0:

        drama_score[person] = (caps / total) * 100

    else:

        drama_score[person] = 0

for person in participants:

    archetypes[person]["DRAMA QUEEN"] = drama_score[person]
```

```python
# The Ghost

ghost_score = {}

for person in participants:

    active_days = set()

    for msg in messages:

        if msg["sender"] == person:

            active_days.add(msg["datetime"].date())

    ghost_score[person] = days - len(active_days)

for person in participants:

    archetypes[person]["GHOST"] = ghost_score[person]
```

```python
# The Group Mom

care_words = [
    "eat","food","sleep","study",
    "take","care","drink",
    "home","safe","okay","ok",
    "fine","medicine","water"
]

mom_score = {}

for person in participants:

    score = 0

    for msg in messages:

        if msg["sender"] == person:

            text = msg["text"].lower()

            for word in care_words:

                if word in text:

                    score += 1

    mom_score[person] = score

for person in participants:

    archetypes[person]["GROUP MOM"] = mom_score[person]
```

```python
# Question Master

question_score = {}

for person in participants:

    count = 0

    for msg in messages:

        if msg["sender"] == person:

            count += msg["text"].count("?")

    question_score[person] = count

for person in participants:

    archetypes[person]["QUESTION MASTER"] = question_score[person]
```

```python
# Comedian

funny_words = [
    "lol","lmao","haha","hehe",
    "😂","🤣","😆","😁"
]

comedy_score = {}

for person in participants:

    score = 0

    for msg in messages:

        if msg["sender"] == person:

            text = msg["text"].lower()

            for word in funny_words:

                score += text.count(word)

    comedy_score[person] = score

for person in participants:

    archetypes[person]["COMEDIAN"] = comedy_score[person]
```

```python
# Choose Best Archetype

final_archetype = {}

for person in participants:

    highest = -1

    best = ""

    for name, score in archetypes[person].items():

        if score > highest:

            highest = score

            best = name

    final_archetype[person] = best
```

```python
# Display Scores

print("="*70)
print("PERSONALITY SCORES")
print("="*70)

for person in sorted(participants):

    print()

    print(person)

    for name, score in archetypes[person].items():

        print(f"{name:<18} {score:.2f}")
```
